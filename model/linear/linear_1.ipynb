```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# langkah 1 : persiapan Data
# Membuat Data dummy untuk regresi linier
np.random.seed(42)
X = np.random.rand(100,1)
y = 3 * X + 2 + np.random.randn(100,1)* 0.1
```

```python
# langkah 2 : Bangun Model 
model = keras.Sequential([
    layers.Dense(1,input_shape=(1,))
])
```

```python
# langkah 3 kompilasi Model 
model.compile(optimizer="adam",loss="mean_squared_error")
```

```python
# langkah 4 latih model
model.fit(X,y,epochs=100,verbose=1)
```

```python
# Langkah 5: Evaluasi Model dan Buat Prediksi
# Evaluasi model dengan data yang sama (hanya untuk contoh ini)
loss = model.evaluate(X, y)
print(f"Loss: {loss}")
```

```python
# Prediksi dengan model yang telah dilatih
preds = model.predict(X)

# Cetak beberapa hasil prediksi
for i in range(5):
    print(f"Input: {X[i][0]}, Ground Truth: {y[i][0]}, Prediction: {preds[i][0]}")
```
