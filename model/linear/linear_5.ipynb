```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import  r2_score , f1_score,confusion_matrix
import pandas as pd
```

```python
# Membuat data sederhana
x_train = np.array([1, 2, 3, 4, 5, 6], dtype=float)  # Jumlah kamar
y_train = np.array([100, 150, 200, 250, 300, 350], dtype=
                   
                   float)  # Harga dalam ribuan dolar

```

```python
# Membuat model Sequential
model = keras.Sequential([
    layers.Dense(units=1, input_shape=[1])
])
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
model.fit(x_train, y_train, epochs=500)
```

```python
print(model.predict(np.array([8.0])))  # Memprediksi harga rumah dengan 7 kamar
```
