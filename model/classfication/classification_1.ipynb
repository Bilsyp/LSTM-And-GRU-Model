```python
import numpy as np 
import pandas as pd

```

```python
df = pd.read_csv("../../data/email.csv")
dfs = df.iloc[:1000]
```

```python
x = dfs["Message"]
y = dfs["Category"]
```

```python
from sklearn.preprocessing  import LabelBinarizer

lb = LabelBinarizer()
y = lb.fit_transform(y).ravel()
lb.classes_
```

```python
from sklearn.model_selection import train_test_split

X_test ,X_train , y_test ,y_train = train_test_split(x,y,random_state=0,train_size=0.24)

print(f'\n x_train:{X_train}\n')
print(f'\n y_train:{y_train}\n')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()

X_train_tfd = vectorizer.fit_transform(X_train)
X_test_tfd = vectorizer.transform(X_test)

# print(f'\n {X_train_tfd}\n')
print(f'\n {X_test_tfd}\n')
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train_tfd,y_train)
```

```python
y_pred = model.predict(X_test_tfd)
y_pred
```

```python
for pred ,sms in zip(y_pred[:5],X_test[:5]):
    print(f"Pred : {pred} sms : {sms}"
          )
```
